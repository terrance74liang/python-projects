```python
start=dt.datetime(2018,9,15)
end=dt.datetime(2019,9,15)
```

```python
# this function sends a request for data and converts it into a dataframe with only its closing price
def get_stock(ticker,start,end):
  data=web.DataReader(f"{ticker}","yahoo",start,end)
  # we are creating a new column named by the ticker with the close column 
  data[f'{ticker}'] = data["Close"]
  # extracting the column and reassigning it to data 
  data = data[[f'{ticker}']] 
  print(data.head(10))
  return data 
```

```python
from functools import reduce
```

```python
# this function combines the chosen stocks into one data frame 
def combine_stocks(tickers,start,end):
  data_frames=[]
  for i in tickers:
    data_frames.append(get_stock(i,start,end))
# reduce method passes second argument's sequence elements into first 
# argument which is a funciton, first two combined, then that result combined with the next
# and so on, merged prints 

#reduce(function,iterable,initializer), initializer is default value for first iteration,
# lambda function takes as argment the value and the element of the reduce function
# value is the next iterable and element is the next element inside the iterable, thus pointer is saved at location since its a generator
# merge(data, intersecting variable, type of merge(merge if intersecting, merge if or, merge first and second coiciding, etc.),...)

# # def reduce(function, iterable, initializer=None):
#     it = iter(iterable)
#     if initializer is None:
#         value = next(it)
#     else:
#         value = initializer
#     for element in it:
#         value = function(value, element)
#     return value

  df_merged=reduce(lambda left,right : pd.merge(left,right,on=['Date'],
                                               how='outer'),data_frames)
  print(df_merged)
  return df_merged
```

```python
# Creating a single dataframe or closing prices for our stocks
stocks = ["SONY", "MDLZ", "BRK-A","PG","INTC","AMGN","GILD","CAT","NEM","CSCO","GEVO","PEP","FCEL","ENB","FSLR","FDX"]
portfolio=combine_stocks(stocks,start,end)
```

```python
from pypfopt.expected_returns import mean_historical_return
from pypfopt.risk_models import CovarianceShrinkage
```

```python
# def mean_historical_return(
#     prices, returns_data=False, compounding=True, frequency=252, log_returns=False
# ):
#     """
#     Calculate annualised mean (daily) historical return from input (daily) asset prices.
#     Use ``compounding`` to toggle between the default geometric mean (CAGR) and the
#     arithmetic mean.
#     :param prices: adjusted closing prices of the asset, each row is a date
#                    and each column is a ticker/id.
#     :type prices: pd.DataFrame
#     :param returns_data: if true, the first argument is returns instead of prices.
#                          These **should not** be log returns.
#     :type returns_data: bool, defaults to False.
#     :param compounding: computes geometric mean returns if True,
#                         arithmetic otherwise, optional.
#     :type compounding: bool, defaults to True
#     :param frequency: number of time periods in a year, defaults to 252 (the number
#                       of trading days in a year)
#     :type frequency: int, optional
#     :param log_returns: whether to compute using log returns
#     :type log_returns: bool, defaults to False
#     :return: annualised mean (daily) return for each asset
#     :rtype: pd.Series
#     """
#     if not isinstance(prices, pd.DataFrame):
#         warnings.warn("prices are not in a dataframe", RuntimeWarning)
#         prices = pd.DataFrame(prices)
#     if returns_data:
#         returns = prices
#     else:
#         returns = returns_from_prices(prices, log_returns)
#     if compounding:
#         return (1 + returns).prod() ** (frequency / returns.count()) - 1
#     else:
#         return returns.mean() * frequency
mu=mean_historical_return(portfolio)
# covariance Shrinkage is a class, we are creating an object of that class and applying the funciton eldoitwolf --> the Ledoit-Wolf shrinkage estimate for a particular
# shrinkage target.
s=CovarianceShrinkage(portfolio).ledoit_wolf()
```

```python
from pypfopt.efficient_frontier import EfficientFrontier

ef = EfficientFrontier(mu,s)
# optimizes for mac sharpe ratio
weights=ef.max_sharpe()

# clean weight and clips zeros
cleaned_weights=ef.clean_weights()
print(dict(cleaned_weights))
print()
ef.portfolio_performance(verbose=True)
```

```python
from pypfopt.discrete_allocation import DiscreteAllocation, get_latest_prices

latest_prices = get_latest_prices(portfolio)

da = DiscreteAllocation(weights, latest_prices, total_portfolio_value=100000)

allocationMeanOpt, leftover = da.greedy_portfolio()
print("Discrete allocation:", allocationMeanOpt)
print("Funds remaining: ${:.2f}".format(leftover))
```

Hierachical risk parity

```python
from pypfopt import HRPOpt

returns = portfolio.pct_change().dropna()
hrp=HRPOpt(returns)
hrp_weights=hrp.optimize()
hrp.portfolio_performance(verbose=True)
print(dict(hrp_weights))
da_hrp=DiscreteAllocation(hrp_weights,latest_prices,total_portfolio_value=100000)
allocationRiskPar,leftover=da_hrp.greedy_portfolio()
print("Discrete allocation (HRP):",allocationRiskPar)
print("Funds remainning (HRP): ${:.2f}".format(leftover))
```

Mean conditional value at risk

```python
from pypfopt.efficient_frontier import EfficientCVaR

s=portfolio.cov()
ef_cvar = EfficientCVaR(mu,s)
cvar_weights=ef_cvar.min_cvar()
cleaned_weights=ef_cvar.clean_weights()
print(dict(cleaned_weights))

da_cvar = DiscreteAllocation(cvar_weights, latest_prices, total_portfolio_value=100000)

allocationCondiValRisk, leftover = da_cvar.greedy_portfolio()
print("Discrete allocation (CVAR):", allocationCondiValRisk)
print("Funds remaining (CVAR): ${:.2f}".format(leftover))
```
